```python
#importing python libraries.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
imdb_df = pd.read_csv('./IMDB-Movie-Data.csv')                     #reading imdb dataset csv file
imdb_df.head(100).style.background_gradient(axis=0, cmap='YlOrRd')   #displaying top 100 datas
```

## Cleaning the dataset 

```python
imdb_df.isnull()  
```

```python
imdb_df.isnull().sum() #checking null valuses in each column
```

```python
imdb_df.isnull().sum().sum()  #checking all NaN or null vaulues.
```

```python
df = imdb_df.dropna().reset_index() #droping all NaN values and reseting the index value.
```

```python
df
```

```python
df.isnull().sum().sum() #checking is there any null value or not
```

```python
df.describe() #Stastical details of the dataset.
```

## Q2.
## what are the top  rated movies :

```python
df_sorted = df.sort_values('Rating',ascending=False) #sorting the DataFrame in decending order according to the movie's rating
```

### Top 10 rating movies and their Director names

```python
top10_movie = df_sorted.head(10)[['Title','Rating','Runtime (Minutes)','Director']].set_index('Title')
top10_movie
```

```python
sns.barplot(x='Rating',y=top10_movie.index,data=top10_movie,hue='Director',dodge=False)
plt.legend(bbox_to_anchor=(1.05,1),loc=2)
```

### Top 10 low rating movies and their Director names

```python
top10_flop = df_sorted.sort_values('Rating',ascending=True).head(10)\
[['Title','Rating','Runtime (Minutes)','Director']].set_index('Title')
top10_flop
```

```python
sns.barplot(x='Rating',y=top10_flop.index,data=top10_flop,hue='Director',dodge=False)
plt.legend(bbox_to_anchor=(1.05,1),loc=2)
```

# Q5 Genre

### how many movies made of each Genre

```python
df.pivot_table(index='Genre', aggfunc='count')[['index']] #displaying the average value of the each Genre.
```

### Display movie details made of Genre is equal to comedy

```python
df[df['Genre'].isin(['Comedy'])]
```

```python
genre_df = df[df['Genre'].isin(['Comedy', 'Action', 'Sci-Fi', 'Drama', 'Horror','Action,Thriller', 'Drama,Thriller', 'Horror,Thriller', 'Crime,Drama,Mystery'])]
```

```python
genre_df['Genre'].value_counts()
```

```python
# plotting how many movies made of each Genre

f = plt.figure()
f.set_figwidth(15)
f.set_figheight(6)

sns.countplot(x='Genre',data=genre_df)
plt.title("Number of Movies of each Genre", fontsize=20)
plt.show()
```

# Q6 voted

### Top voted movies

```python
top_voted_movies = df.nlargest(5, 'Votes')[['Title','Votes','Rating']].set_index('Title')  #top 5 voted movies Name and their Ratings
```

```python
top_voted_movies
```

```python
sns.barplot(x='Rating',y=top_voted_movies.index, data=top_voted_movies, hue='Votes',dodge=False)
plt.title("Top voted movies and their Ratings", fontsize=20)
plt.legend(bbox_to_anchor=(1.05,1),loc=2)
```

## pie plot of top rated movies and their votes percentages

```python
colors=['b','c','r','y','g']
explode = [0.03,0,0,0,0]
textprops={'fontsize':12}
plt.pie(top_voted_movies.Votes, labels=top_voted_movies.index,colors=colors, explode=explode,autopct='%0.2f%%',startangle=90,textprops=textprops)
plt.show()
```
